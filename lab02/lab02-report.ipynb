# Exercise 1:

```python
import timeit
import matplotlib.pyplot as plt

#a test to verify that get item and update item are O(1) for dictionaries

graphx = []
graphy = []

ugraphx = []
ugraphy = []

counter = 0
d1 = {i: "potato" for i in range(5000)}

runtime1 = 0
runtime2 = 0
d1keys = d1.keys()
# loop for get test
for x in d1keys:
    t = timeit.Timer("d1.get(x)", "from __main__ import d1,x")
    graphy.append(t.timeit(number = 5000))
    runtime1 += graphy[counter]
    graphx.append(counter)
    counter = counter + 1
    
counter = 0
#loop for update test
for x in d1keys:
    t = timeit.Timer("d1.update({x:'baked'})", "from __main__ import d1,x")
    ugraphy.append(t.timeit(number = 5000))
    runtime2 += ugraphy[counter]
    ugraphx.append(counter)
    counter = counter + 1
    


#graph for get
plt.title("Get")
plt.scatter(graphx,graphy)
plt.xlabel("Runs")
plt.ylabel("Time in seconds")
plt.show()
print("Total Runtime: %.5f Average Runtime: %.5f" %(runtime1, runtime1/5000))

#graph for update
plt.title("update")
plt.scatter(ugraphx,ugraphy)
plt.xlabel("Runs")
plt.ylabel("Time in seconds")
plt.show()
print("Total Runtime: %.5f Average Runtime: %.5f" %(runtime2, runtime2/5000))
```

### Discussion of the results
The results show that both get and update have a graph that is big O(1). It also shows that updating values in a dictionary requires slightly more time than in a list. The extra runtime is likely from the time is takes to match the key before the value is updated.
